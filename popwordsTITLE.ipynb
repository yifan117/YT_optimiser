```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from collections import Counter

gbdf = pd.read_csv("./GB_youtube_trending_data.csv")
usdf = pd.read_csv("./US_youtube_trending_data.csv")
```

```python
Counter(" ".join(usdf["title"]).split()).most_common(2000)
```

```python
Counter(" ".join(gbdf["title"]).split()).most_common(2000)
```

```python
plotdata = pd.DataFrame ({
    
    "GB":[46734, 30478, 31076, 12704, 10407, 10284],
    "US":[36333, 36373, 26504, 12237, 9484, 10675]
    }, index = ["|", "-", "The", "I", "&", "In"]
)
plotdata.head()

plotdata.reset_index().plot(
    x = "index", y = ["US", "GB"], kind = "bar"
)
plt.title("Comparison of Most Popular Title Keywords in GB and US (UNFILTERED)")
plt.xlabel("Keyword")
plt.ylabel("Frequency of Use")

# unfiltered graph
```
