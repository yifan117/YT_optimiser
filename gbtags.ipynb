```python
import pandas as pd
import numpy as np

df = pd.read_csv("./GB_youtube_trending_data.csv")
```

```python
df.sort_values(
    by="view_count",
    ascending = False
)[["title", "view_count","tags"]].head(100)
```

```python
from collections import Counter
Counter(" ".join(df["tags"]).split()).most_common(2000)
```

```python
pd.Series(' '.join(df['tags']).lower().split()).value_counts()[:100]
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

text2 = " ".join(df['tags'])

word_cloud2 = WordCloud(width = 4500, height = 3000, random_state = 42, collocations = False, background_color = 'black', colormap = 'Pastel1').generate(text2)
```

```python
plt.imshow(word_cloud2, interpolation='bilinear')
plt.axis("off")
plt.show()
```
