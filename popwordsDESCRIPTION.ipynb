```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from collections import Counter

gbdf = pd.read_csv("./GB_youtube_trending_data.csv")
usdf = pd.read_csv("./US_youtube_trending_data.csv")
```

```python
Counter(" ".join(map(str, usdf["description"])).split()).most_common(2000)
```

```python
Counter(" ".join(map(str, gbdf["description"])).split()).most_common(2000)
```

```python
plotdata = pd.DataFrame ({
    
    "GB":[462023, 320232, 315834, 241278, 198080, 189154],
    "US":[449309, 315822, 320410, 213007, 213248, 195612]
    }, index = ["The", "To", "And", "-", "A", "Of"]
)
plotdata.head()

plotdata.reset_index().plot(
    x = "index", y = ["US", "GB"], kind = "bar"
)
plt.title("Comparison of Most Popular Description Keywords in GB and US (UNFILTERED)")
plt.xlabel("Keyword")
plt.ylabel("Frequency of Use")

# unfiltered graph
```
